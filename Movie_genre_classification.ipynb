```python
print("Columns in train_data:", train_data.columns)
print("Columns in test_data:", test_data.columns)
print("columns in test_data_solution:", test_data_solution )

import re
def clean_text(text):
    text = re.sub(r'[^a-zA-Z0-9]', ' ', text)
    text = re.sub(r'\s+', ' ',text)
    text = re.sub(r'^\s+/\s+', ' ', text)
    text = re.sub(r'\s+$', ' ', text)
    text = re.sub(r'http\s+://s+', ' ', text)
    text = text.lower().strip()
    return text

train_data['Text_cleaning'] = train_data['Description'].apply(clean_text)
test_data['Text_cleaning'] = test_data['Description'].apply(clean_text)
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfVectorizer = TfidfVectorizer()
x_train = tfidfVectorizer.fit_transform(train_data['Text_cleaning'])
x_test = tfidfVectorizer.transform(test_data['Text_cleaning'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(train_data['Description'], train_data['Genre'], test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x_train_vectorized = vectorizer.fit_transform(x_train)
x_test_vectorized = vectorizer.transform(x_test)
print(x_train_vectorized)
print(x_test_vectorized)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train_vectorized, y_train)
```

```python
model.score(x_test_vectorized, y_test)
```

```python
model.predict(x_test_vectorized)
```

```python
from sklearn.naive_bayes import MultinomialNB
m1 = MultinomialNB()
m1.fit(x_train_vectorized, y_train)
```

```python
m1.score(x_test_vectorized, y_test)
```

```python
m1.predict(x_test_vectorized)
```

```python
y_pred = m1.predict(x_test_vectorized)
from sklearn.metrics import accuracy_score
Accuracy = accuracy_score(y_test,y_pred)
Accuracy
```
