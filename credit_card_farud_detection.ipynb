```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fraudtest_data = pd.read_csv('fraudTest.csv')
fraudtest_data.head()
```

```python
fraudtrain_data = pd.read_csv('fraudTrain.csv')
fraudtrain_data.head()
```

```python
fraudtrain_data.info()
```

```python
fraudtest_data.info()
```

```python
fd = pd.concat([fraudtrain_data, fraudtest_data])
fd.head()
```

```python
fd.isnull().sum()
```

```python
fd.shape[0]
```

```python
fd.fillna(0, inplace=True)
fd.isnull().sum()
```

```python
fd.drop(columns=['Unnamed: 0', 'gender', 'street', 'job', 'unix_time','trans_num','merchant','zip'], inplace=True)
fd.head()
```

```python
fd.duplicated().sum()
```

```python
fd.shape[0]
```

```python
fd.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

columns = ['trans_date_trans_time','category','first','last','city','state','lat','long','dob','merch_lat','merch_long']
for col in columns:

    if fd[col].apply(type).nunique() > 1:
        fd[col] = fd[col].astype(str)
    fd[col] = le.fit_transform(fd[col])
```

```python
fd.head()
```

```python
x_train = fd.drop(columns=['is_fraud'])
y_train = fd['is_fraud']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=40)
rf.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
rf.score(x_test,y_test)
```

```python
dt.score(x_test,y_test)
```

```python
svm.score(x_test,y_test)
```

```python
y_predictlr = lr.predict(x_test)
y_predictlr
```

```python
y_predictrf = rf.predict(x_test)
y_predictrf
```

```python
y_predictdt = dt.predict(x_test)
y_predictdt
```

```python
y_predictsvm = svm.predict(x_test)
y_predictsvm
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm1 = confusion_matrix(y_test, y_predictlr)
cm2 = confusion_matrix(y_test, y_predictrf)
cm3 = confusion_matrix(y_test, y_predictdt)
cm4 = confusion_matrix(y_test, y_predictsvm)
```

```python
cm1_accuracy = accuracy_score(y_test, y_predictlr)
cm2_accuracy = accuracy_score(y_test, y_predictrf)
cm3_accuracy = accuracy_score(y_test, y_predictdt)
cm4_accuracy = accuracy_score(y_test, y_predictsvm)
```

```python
print(cm1_accuracy)
print(cm2_accuracy)
print(cm3_accuracy)
print(cm4_accuracy)
```

```python
print(cm1)
print(cm2)
print(cm3)
print(cm4)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predictlr))
```
